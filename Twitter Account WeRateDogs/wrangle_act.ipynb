```python
import pandas as pd
import numpy as np
import requests
import json
import tweepy
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

# Gathering Data for this Project

```python
# Import the Twitter_Archive data
twitter_archive = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
# Download the file from Udacity Server
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
with open("image_predictions.tsv", mode = 'wb') as f:
    f.write(response.content)
image_predictions = pd.read_csv('image_predictions.tsv',sep='\t')
```

```python
# Authenticate to Twitter
consumer_key= ""
consumer_secret = ""
access_token_key = ""
access_token_secret = ""

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token_key, access_token_secret)

# Create API object
api = tweepy.API(auth)
```

```python
#Query the Twitter API for each tweet's JSON data using Python's 
#Tweepy library and store each tweet's entire set of JSON data in a 
#file called tweet_json.txt file. Each tweet's JSON data should be 
#written to its own line. 
tweet_ids = twitter_archive.tweet_id.unique()
with open('tweet_json.txt', 'w') as outfile:
    for tweet_id in tweet_ids:
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except Exception as e:
            print(e)
```

```python
#Then read this .txt file line by line into a pandas DataFrame with 
#(at minimum) tweet ID, retweet count, and favorite count. 
df_list = []
with open('tweet_json.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        try:
            tweets_line = json.loads(line)

            tweet_id = tweets_line['id']
            full_text = tweets_line['full_text']
            media_url = tweets_line['entities']['media'][0]['media_url_https'] 
            url = tweets_line['entities']['media'][0]['url'] 
            rt_count = tweets_line['retweet_count']
            fav_count = tweets_line['favorite_count']
            df_list.append({'tweet_id': tweet_id, 
                         "full_text":full_text,
                         "media_url":media_url,
                         "url":url,
                         'rt_count': rt_count, 
                         'fav_count': fav_count})
        except Exception as e:
            print(e)
df = pd.DataFrame(df_list, columns = ['tweet_id', 'full_text','media_url','url','rt_count','fav_count'])
```

### twitter_archive table

- The dog stages(pupper, poppo, doggo) should in one column

Define: Use .melt() to melt these three columns in one.

Clean:

```python
twitter_archive_copy = twitter_archive_copy.query('doggo == "doggo" or pupper == "pupper" or puppo == "puppo"')
stage_cols = ['doggo', 'pupper', 'puppo']
non_stage_cols = twitter_archive_copy.columns.difference(stage_cols)
twitter_archive_copy = twitter_archive_copy.melt(id_vars = non_stage_cols, value_vars = stage_cols, var_name = 'stage_level', value_name = 'stage').dropna(subset = ['stage'])
twitter_archive_copy = twitter_archive_copy.drop(labels = ['stage_level'], axis=1)
```

Test:

```python
twitter_archive_copy.info()
```

Clean:

### Insight 1 : Does any correlation exsit between the ratings and the dog ages?

```python
twitter_archive_master['rating'] = twitter_archive_master['rating_numerator']/twitter_archive_master['rating_denominator']
twitter_archive_master.head(1)
```

About this insight, I was trying to find if there's some correlation between the dog age and the dog rating for these twitters.

```python
sb.boxplot(data = twitter_archive_master, x = 'stage', y = 'rating');
plt.ylim(0.75, 1.5)
plt.title('Dog Rating VS Ages');
```

In the three stages, the ratings of the poppo stage(between old and young) are intensively distributed than the doggo and pupper; the pupper stage(the young) gets a same box range but the puppo has no bottom whisker which means it has less lower ratings. The doggo stage(the adult) gets a relatively lower ratings than pupper and poppo, and the dirtribution is much more scattered and has very low outliers, like 0.8.

### Insight 3 : Does any correlation exsit between the ratings and the favorite counts of the tweets?

About this insight, I was trying to find if there's some correlation between the favorite counts and the dog ratings for these twitters.

```python
plt.scatter(data = twitter_archive_master, y = 'rating', x = 'fav_count', \
            alpha = 1/5);
plt.ylim(0.75, 1.5);
plt.xlabel('Fav_count');
plt.ylabel('Rating');
plt.title('Dog fav_count VS Rating');
```
